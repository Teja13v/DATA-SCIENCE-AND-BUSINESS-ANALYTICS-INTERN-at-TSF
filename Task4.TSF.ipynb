### Import Libraries and Dataset

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
df=pd.read_csv("gt.csv")
df.head(5)
```

```python
df= df[['Year','Month','Extended','Day','Country','State','Region','City','Latitude','Longitude',
                 'AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type']]
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

### Top 10 Countries with most Terrorist Attacks

```python
print("Country with the least attacks:\n",df['Country'].value_counts().head(5))
plt.figure(figsize = (10,5))
sns.barplot(df['Country'].value_counts()[:15].index,df['Country'].value_counts()[:15].values)
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

### States with most Terrorist Attack

```python
plt.figure(figsize = (10,5))
sns.barplot(df['State'].value_counts()[:15].index,df['State'].value_counts()[:15].values)
plt.title('Top States Affected')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

### Regions with most Terrorist Attacks

```python
plt.figure(figsize = (10,5))
sns.barplot(df['Region'].value_counts()[:15].index,df['Region'].value_counts()[:15].values)
plt.title('Top Regions Affected')
plt.xlabel('Regions')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

### Cities with most Terrorist Attacks

```python
plt.figure(figsize = (10,5))
sns.barplot(df['City'].value_counts()[:15].index,df['City'].value_counts()[:15].values)
plt.title('Top City Affected')
plt.xlabel('Cities')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

### Most Frequent Attack Type

```python
plt.figure(figsize = (10,5))
sns.barplot(df['AttackType'].value_counts()[:15].index,df['AttackType'].value_counts()[:15].values)
plt.title('Most Frequent Attack Tyoe')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

### Most frequent Weapon used

```python
plt.figure(figsize = (15,7))
sns.barplot(df['Weapon_type'].value_counts()[:15].index,df['Weapon_type'].value_counts()[:15].values)
plt.title('Most popular Weapon Type')
plt.xlabel('Weapon')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

### Killed vs Wounded

```python
df.plot(kind = 'scatter', x = 'Killed', y = 'Wounded', alpha = 0.5, color = 'blue', figsize = (10,5))
plt.xlabel('Killed')
plt.ylabel('Wounded')
plt.title('Killed vs Wounded')
plt.show()
```
